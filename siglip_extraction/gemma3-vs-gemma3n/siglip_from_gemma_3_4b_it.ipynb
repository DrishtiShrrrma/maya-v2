```python
from transformers import Gemma3ForConditionalGeneration
model_id = "google/gemma-3-4b-it"

model = Gemma3ForConditionalGeneration.from_pretrained(model_id)



print(model)
```

```python
siglip = model.vision_tower
print(siglip)


```
